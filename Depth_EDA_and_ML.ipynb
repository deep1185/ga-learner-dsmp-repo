# ML Classification

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import ExtraTreeClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
```

```python
rc=RandomForestClassifier()
lr=LogisticRegression()
dc=DecisionTreeClassifier()
ec=ExtraTreeClassifier()
esc=ExtraTreesClassifier()
gbc=GradientBoostingClassifier()
```

```python
model_dict={0:(rc,'RandomForestClassifier'),
            1:(lr,'LogisticRegression'),
            2:(dc,'DecisionTreeClassifier'),
            3:(ec,'ExtraTreeClassifier'),
            4:(esc,'ExtraTreesClassifier'),
            5:(gbc,'GradientBoostingClassifier')}
```

```python
def model_fitting(x_train,x_test,y_train,y_test):
    result=[]
    for k in model_dict.keys():
        result_dict={}
        result_dict['Model']=model_dict[k][1]
        model=model_dict[k][0]
        model.fit(x_train,y_train)
        y_pred=model.predict(x_test)
        result_dict['Precision Score']=precision_score(y_test,y_pred)
        result_dict['Recall Score']=recall_score(y_test,y_pred)
        result_dict['Accuracy Score']=accuracy_score(y_test,y_pred)
        result_dict['F1 Score']=f1_score(y_test,y_pred)
        result.append(result_dict)
        #print(model_dict[k][1],classification_report(y_test,y_pred),sep='\n')
    return pd.DataFrame(result)
```

```python
classification_data=pd.get_dummies(data,columns=['Summary','month','time'],drop_first=True)
```

```python
classification_data=pd.get_dummies(classification_data,columns=['Precip Type'])
```

```python
x=classification_data.drop(columns=['Formatted Date','Apparent Temperature (C)','Daily Summary',\
                                    'Precip Type_Other','Precip Type_rain','Precip Type_snow'])
y=classification_data['Precip Type_rain']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.7,random_state=7)
```

```python
model_fitting(x_train,x_test,y_train,y_test)
```
