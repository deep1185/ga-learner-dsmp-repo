```python
filename = 'Deep.xlsx'
df = pd.read_excel(filename)
df.head()
```

```python
df.fillna(0, inplace=True)
df.head()
```

```python
df1 = df[df['Pareto ']==1.]
df1.head()
```

```python
df0 = df[df['Pareto ']==0.]
df0.head()
```

```python
wb1 = df1.filter(['Super Category', 'Brand', 'Pareto ', 'Type', '2023 Wk 48Sales', '2023 Wk 48RUSales', '2023 Wk 49Sales', '2023 Wk 49RUSales', '2023 Wk 50Sales', '2023 Wk 50RUSales', '2023 Wk 51Sales', '2023 Wk 51RUSales'], axis=1)
wb1.head()
```

```python
wb1.rename(columns= {'2023 Wk 48Sales':'Wk 48Sales',
                                               '2023 Wk 49Sales':'Wk 49Sales',
                                               '2023 Wk 50Sales':'Wk 50Sales',
                                               '2023 Wk 51Sales':'Wk 51Sales',
                                               '2023 Wk 48RUSales':'Wk 48RUSales',
                                               '2023 Wk 49RUSales':'Wk 49RUSales',
                                               '2023 Wk 50RUSales':'Wk 50RUSales',
                                               '2023 Wk 51RUSales':'Wk 51RUSales'}, inplace=True)

wb1
```

```python
wb1_weekly_categorywise_sales = wb1.groupby('Super Category').agg({'Wk 48Sales': np.sum,
                                                                   'Wk 48RUSales': np.sum,
                                                                   'Wk 49Sales': np.sum,
                                                                   'Wk 49RUSales': np.sum,
                                                                   'Wk 50Sales': np.sum,
                                                                   'Wk 50RUSales': np.sum,
                                                                   'Wk 51Sales': np.sum,
                                                                   'Wk 51RUSales': np.sum})
```

```python
wb1_weekly_categorywise_sales.reset_index(inplace = True)
```

```python
wb1_weekly_categorywise_sales['Wk 48 RU Per'] = np.round(wb1_weekly_categorywise_sales['Wk 48RUSales'] / wb1_weekly_categorywise_sales['Wk 48Sales']*100, 2)
wb1_weekly_categorywise_sales['Wk 49 RU Per'] = np.round(wb1_weekly_categorywise_sales['Wk 49RUSales'] / wb1_weekly_categorywise_sales['Wk 49Sales']*100, 2)
wb1_weekly_categorywise_sales['Wk 50 RU Per'] = np.round(wb1_weekly_categorywise_sales['Wk 50RUSales'] / wb1_weekly_categorywise_sales['Wk 50Sales']*100, 2)
wb1_weekly_categorywise_sales['Wk 51 RU Per'] = np.round(wb1_weekly_categorywise_sales['Wk 51RUSales'] / wb1_weekly_categorywise_sales['Wk 51Sales']*100, 2)
```

```python
wb1_weekly_categorywise_sales
```

**Super Categories performing best in terms of Sales & RU Sales (Pareto-1)**

```python
sup_cat_high_sales = wb1_weekly_categorywise_sales.idxmax(axis = 0, numeric_only = True)
super_category_highest_sales = pd.Series(sup_cat_high_sales).to_frame().reset_index().rename(columns = {'index': 'Week of Sales', 0: 'Best Super Category'})
super_category_highest_sales['Best Super Category'] = super_category_highest_sales['Best Super Category'].apply(lambda x: wb1_weekly_categorywise_sales.loc[x, 'Super Category'])
super_category_highest_sales
```

```python
wb1_category_sales = wb1_weekly_categorywise_sales.filter(['Super Category', 'Wk 48 RU Per', 'Wk 49 RU Per', 'Wk 50 RU Per', 'Wk 51 RU Per'])
melted_wb11 = pd.melt(wb1_category_sales, id_vars=['Super Category'], var_name='Weekly Sales RU Per', value_name='Value')
```

```python
plt.figure(figsize = (20, 15))
sns.barplot(x='Value', y='Super Category', hue='Weekly Sales RU Per', data=melted_wb11, orient='h')
```

```python
wb1_weekly_brandwise_sales = wb1.groupby('Brand').agg({'Wk 48Sales': np.sum,
                                                                   'Wk 48RUSales': np.sum,
                                                                   'Wk 49Sales': np.sum,
                                                                   'Wk 49RUSales': np.sum,
                                                                   'Wk 50Sales': np.sum,
                                                                   'Wk 50RUSales': np.sum,
                                                                   'Wk 51Sales': np.sum,
                                                                   'Wk 51RUSales': np.sum})
```

```python
wb1_weekly_brandwise_sales.reset_index(inplace = True)
```

```python
wb1_weekly_brandwise_sales['diff_RUSales'] = wb1_weekly_brandwise_sales['Wk 51RUSales']-wb1_weekly_brandwise_sales['Wk 48RUSales']
```

**Fort Collins only Brand having postive trend of RUSales from week 48 to week 51 for Pareto 1 (Waterfall for Pareto-1)**

```python
wb1_weekly_brandwise_sales.sort_values(by='diff_RUSales',ascending=False)
```

```python
wb1_weekly_brandwise_sales['Wk 48 RU Per'] = np.round(wb1_weekly_brandwise_sales['Wk 48RUSales'] / wb1_weekly_brandwise_sales['Wk 48Sales']*100, 2)
wb1_weekly_brandwise_sales['Wk 49 RU Per'] = np.round(wb1_weekly_brandwise_sales['Wk 49RUSales'] / wb1_weekly_brandwise_sales['Wk 49Sales']*100, 2)
wb1_weekly_brandwise_sales['Wk 50 RU Per'] = np.round(wb1_weekly_brandwise_sales['Wk 50RUSales'] / wb1_weekly_brandwise_sales['Wk 50Sales']*100, 2)
wb1_weekly_brandwise_sales['Wk 51 RU Per'] = np.round(wb1_weekly_brandwise_sales['Wk 51RUSales'] / wb1_weekly_brandwise_sales['Wk 51Sales']*100, 2)
```

```python
wb1_brand_sales = wb1_weekly_brandwise_sales.filter(['Brand', 'Wk 48 RU Per', 'Wk 49 RU Per', 'Wk 50 RU Per', 'Wk 51 RU Per'])
melted_wb12 = pd.melt(wb1_brand_sales, id_vars=['Brand'], var_name='Weekly Sales RU Per', value_name='Value')
```

```python
wb1_brand_sales
```
