```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from scipy.stats import shapiro,normaltest
import seaborn as sns; sns.set()
import statsmodels.api as sm
from statsmodels.graphics.gofplots import qqplot
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv("F:\\Python work\\titanic\\weatherHistory.csv")
data
```

```python
data.drop(columns=['Apparent Temperature (C)','Loud Cover','Formatted Date','Summary','Daily Summary'],inplace=True)
data
```

```python
df=pd.get_dummies(data,columns=['Precip Type'])
df
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df.drop(columns=['Precip Type_snow'],inplace=True)
df
```

```python
rfr=RandomForestClassifier()
lr=LogisticRegression()
```

```python
exp_tup={0:(lr,'LogisticRegressor'),1:(rfr,'RandomForestClassifier')}
```

```python
x=df.drop('Precip Type_rain',axis=1)
```

```python
y=df['Precip Type_rain']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=7)
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, f1_score
```

```python
res_ls = []
for k in exp_tup.keys():
    model = exp_tup[k][0]
    name = exp_tup[k][1]
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    accuracy = accuracy_score(y_test, y_pred)
    res_ls.append((name, precision,recall,accuracy))
res_df = pd.DataFrame(res_ls)
res_df.columns = ['model', 'precision','recall','accuracy']
res_df.sort_values('precision', inplace=True)
```

```python
result=pd.DataFrame(res_df)
```
