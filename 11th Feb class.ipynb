```python
plt.plot(x, np.sin(x-0), color = 'blue')
plt.ylim(-1.5, 1.5)
plt.xlim(1, 15)
plt.grid() #to make plot appear on number system to see which value is plotted where
plt.axis('equal') # to make plot cover equal spaces on both sides, range of plot will change to make plot look symmetrical and look good
plt.show()
```

```python
plt.plot(x, np.sin(x-0), color = 'blue')
plt.ylim(-1.5, 1.5)
plt.xlim(1, 15)
plt.grid()
plt.axis('equal')
plt.title("This is sine plot to understand syntax basics") #title
plt.show()
```

```python
plt.plot(x, np.sin(x-0), color = 'blue')
plt.ylim(-1.5, 1.5)
plt.xlim(1, 15)
plt.grid()
plt.axis('equal')
plt.xlabel("x axis")# name given to x axis
plt.ylabel("y axis")# name given to y axis
plt.title("This is sine plot to understand syntax basics") #title given
plt.show()
```

```python
plt.plot(x, np.sin(x-0), color = 'blue', label = 'Sinx')  # naming and coloring graph line
plt.plot(x, np.cos(x-0), color = 'red', label = 'Cosx') # naming and coloring graph line
plt.ylim(-1.5, 1.5)
plt.xlim(1, 15)
plt.grid()
plt.title("This is sin plot before real dataset execution to understand basic syntax")
plt.xlabel("x axis")
plt.ylabel("y axis")
# plt.axis('equal'), now not needed so commented out
plt.legend() #it puts all labels put in syntax to execution and shows all in the plot and not outside
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.figure(figsize = (6, 2))
plt.scatter(x, y, c = 'r', alpha = .9, marker = "*" ) # "c" =color parameter, alpha= color intensity (0-1), scatter plot shows how value pairs of x and y are scattered
plt.xlabel("x")
plt.ylabel("y")
plt.title("scatter plot")
plt.show()
```

```python
x=["a", "b", "c", "d"]#categories
y=[4, 6, 1, 2]#counts
```

```python
#pie chart

langs = ['C', 'Python', 'Java']
students = [20, 100, 40]

plt.pie(students, labels = langs, autopct = '%1.1f%%')
plt.title("Pie chart")
plt.show()
```

```python
my_explode = (0.0, 0.1, 0.1)

langs = ['C', 'Python', 'Java']
students = [20, 100, 40]

plt.pie(students, labels = langs, autopct = '%1.1f%%', explode = my_explode, shadow =  True)
plt.title("Pie chart")
plt.show()
```

```python
#3d plot
x = np.random.rand(20)
y = np.random.rand(20)
z = np.random.rand(20)
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.scatter (x,y,z,c = '#FF5733', alpha = 0.9)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.show()
```

```python
import seaborn as sns
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.get_dataset_names()
```

```python
sns.histplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill']) #shows probability density , kde(kernel density estimator)
```

```python
sns.distplot(tips['total_bill'], kde=False, bins =30) #if kde line is unwanted
```

```python
sns.scatterplot(x=tips.total_bill, y=tips.tip) #relation between numerical variables total_bll & tip
#as we can see highest tip is paid by the person who has paid highest total_bill
```

```python
#smoker and non-smoker profile
sns.relplot(x=tips.total_bill, y=tips.tip, data=tips, hue='smoker') #hue is a parameter, here we want to see smoker profile
```

```python
sns.relplot(x=tips.total_bill, y=tips.tip, data=tips, hue='size')
```

```python
sns.relplot(x=tips.total_bill, y=tips.tip, data=tips, style='size') #instead of color we distinguish by style of plotting of parameter 'size'
```

```python
sns.relplot(x=tips.total_bill, y=tips.tip, data=tips, style='smoker')
```

```python
sns.relplot(x=tips.total_bill, y=tips.tip, data=tips, style='size', hue='time') # maximum tips paid in dinner time that too by smokers
#style is like a marker, hue is like a groupby
```

```python
sns.catplot(x='day', y='total_bill', data=tips) #category wise plotting, better substitute of barplot to show category wise data
#saturday and sunday has more crowd, max bill happens on saturday
```

```python
sns.barplot(x='day', y='total_bill', data=tips, color='#FF5733')
```

```python
sns.jointplot(x=tips.total_bill, y=tips.tip, data=tips, kind="reg") #shows scatter and distribution of individual variables, "reg" for understanding reltionship better
```

```python
sns.jointplot(x=tips.total_bill, y=tips.tip, data=tips, kind="hex")
```

```python
sns.jointplot(x=tips.total_bill, y=tips.tip, data=tips, kind="kde") #contour plot, use cases in weather figures and plots and seismographs(earthquake)
```

```python
#pairplot plots numerical variables relation pairwise from the dataset
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex') #includes categorical data as parameter
```

```python
sns.pairplot(tips,hue='smoker')
```

```python
sns.pairplot(tips,hue='smoker', palette='coolwarm')
```

```python
#boxplot is used to determine outliers
sns.boxplot(x='day', y='total_bill', data=tips, palette = 'rainbow')
```

```python
sns.boxplot(data=tips, palette = 'rainbow', orient='h')
```

```python
#correlation >> to understand the relationship between
tips_num = tips[['total_bill', 'tip', 'size']]
tips_num.corr()
```

```python
sns.heatmap(tips_num.corr(), annot = True, cmap = 'coolwarm') #annot shows correlation
```

```python
sns.countplot(x = 'sex', data=tips) #categorical variables
```

```python
sns.lmplot(x = 'total_bill', y='tip', data = tips, height = 20, aspect = 1)
```

```python
sns.catplot(x = 'day', y = 'tip', data = tips, kind='swarm') #swarm plot
```
