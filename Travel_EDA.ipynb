```python
# Extracting categorical and numerical columns

cat_cols = [col for col in df.columns if df[col].dtype == 'object']
num_cols = [col for col in df.columns if df[col].dtype != 'object']
```

```python
cat_cols
```

```python
for column in cat_cols:
  plt.figure(figsize=(8,5))
  sns.countplot(x= column, data=df, palette='viridis')
  plt.title("Univariate analysis of Categorical columns")
  plt.legend()
  plt.show()
```

```python
for column in num_cols:
  plt.figure(figsize=(8,5))
  sns.histplot(df[column], kde=True)
  plt.title("Univariate analysis of Numerical Columns")
  plt.legend()
  plt.show()
```

```python
cross_tab=pd.crosstab(df['MaritalStatus'], df['ProdTaken'], normalize= 'index') #normalize=index means % of a column in respect to another, #normalize=True means % out of total
cross_tab
```

```python
cross_tab.plot(kind='bar', stacked=True, color=['skyblue', 'lightcoral'], figsize= (10,6))
plt.title("stacked bar plot of marital statusand prod taken")
plt.xlabel("Marital Staus")
plt.ylabel("percentage")
plt.show()
```
